```python
grades = let
    tmpdf = DataFrame(:user_id => UInt8[], :score => UInt8[])
    filename = "../Data/labgrade.txt"
    lines = readlines(filename)
    for line in lines
        halves = split(line)
        id = parse(UInt8, halves[1])
        score = parse(UInt8, halves[2])
        push!(tmpdf, (id, score))
    end
    tmpdf
end
```
